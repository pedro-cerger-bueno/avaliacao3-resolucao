**Resto de divisão: Caracterizado por sua simplicidade e alto nível de colisões. Usado para distribuição uniforme de elementos em uma tabela hash**

**Multiplicação: Multiplica a chave e extrai parte do resultado. Diferentemente do hash por resto de divisão, essa técnica não é tão dependente no tamanho da tabela**

**Dobramento: Técnica em que a chave é dividida em partes menores que então são combinadas para formar o índice na tabela hash**
